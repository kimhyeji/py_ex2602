```python
# 클래스 -> 객체를 표현하기 위한 문법의 일종,
# 프로그래밍으로 객체를 만들 때 사용하는 것

# 전사, 마법사, 궁수, 자동차, 비행기 -> object
# -> 현실, 가상, 개념

# 속성(명사적) - attribute
# 능력(기능, 동사적) - method


class Person:
    def hello(self):
        print("hi")

jin = Person() # jin은 Person의 인스턴스
jin.hello()
```
